```python
def crossover(sp, p1, p2):
    def eval(route, idx):
        v1, v2, v3, v4 = route[idx - 1][0][1], route[idx][0][0], route[idx][0][1], route[idx + 1][0][0]
        return sp[v1, v2] + sp[v3, v4] - sp[v2, v3]

    r1, r2 = choice(p1), choice(p2)
    a, b = randint(len(p1) - 1), randint(len(p2) - 1)
    old_route, new_route = p1[a], p2[b]
    p0 = p1[:a] + p1[a + 1:]  # 先不放new_route，防止下面找的时候有问题

    # ! Remove duplicated tasks and insert unserved tasks
    old_route, new_route = set(old_route), set(new_route)
    common = old_route & new_route
    unserved = old_route - common
    duplicated = new_route - common

    # Remove duplicated tasks
    # find the position of the duplicate task
    for task in unserved:
        for route in p0:
            if task in route:
                idx1 = route.index(task)
                

                # 判断应该移除哪一个
                s1 = sp[route[idx1 - 1][0][1], route[idx1][0][0]] + \
                     sp[route[idx1][0][1], route[idx1 + 1][0][0]] - \
                     sp[route[idx1 - 1][0][1], route[idx1 + 1][0][0]]



    return p0
```
