```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# ignore warnings
# import warnings
# warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../data/adult.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df[df == "?"].count()
```

```python
# workclass, occupation and native.country contain missing values
df[df == "?"] = np.nan
df.info()
```

```python
# 用众数填充
for col in ["workclass", "occupation", "native.country"]:
    df.fillna(df[col].mode(0)[0], inplace=True)

df.isnull().sum()
```

```python
X = df.drop(["income"], axis=1)
y = df["income"]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape
```

### Feature Engineering

```python
# Encode categorical variables
from sklearn import preprocessing
categorical = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
for feature in categorical:
    le = preprocessing.LabelEncoder()
    X_train[feature] = le.fit_transform(X_train[feature])
    X_test[feature] = le.fit_transform(X_test[feature])
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.fit_transform(X_test), columns=X.columns)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print("Logistic Regression accuracy with all features: {}".format(accuracy_score(y_test, y_pred)))
```

#### Logistic Regression with PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
X_train = pca.fit_transform(X_train)
pca.explained_variance_ratio_
```

```python
# Logistic Regression with first 13 features
X = df.drop(['income', 'native.country'], axis=1)
y = df['income']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

categorical = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex']
for feature in categorical:
    le = preprocessing.LabelEncoder()
    X_train[feature] = le.fit_transform(X_train[feature])
    X_test[feature] = le.transform(X_test[feature])

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print('Logistic Regression accuracy with the first 13 features: {}'.format(accuracy_score(y_test, y_pred)))


```

```python
X = df.drop(['income','native.country', 'hours.per.week'], axis=1)
y = df['income']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)


categorical = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex']
for feature in categorical:
        le = preprocessing.LabelEncoder()
        X_train[feature] = le.fit_transform(X_train[feature])
        X_test[feature] = le.transform(X_test[feature])


X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X.columns)

X_test = pd.DataFrame(scaler.transform(X_test), columns = X.columns)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print('Logistic Regression accuracy with the first 12 features: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
